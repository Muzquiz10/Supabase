```python
import dotenv
import psycopg2
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()

OUTPUT_DIR = "./export_supabase/vtc_test_BBDD"
os.makedirs(OUTPUT_DIR, exist_ok=True)
SCHEMA = os.getenv("SCHEMA")

# --- CONEXIÓN ---
conn = psycopg2.connect(
host = os.getenv("HOST"),
port = os.getenv("PORT"),
database = os.getenv("DB"),
user = os.getenv("USER"),
password = os.getenv("PASSWORD")
)

cur = conn.cursor()

# 1. Obtener todas las tablas del schema
cur.execute(f"""
    SELECT table_name 
    FROM information_schema.tables 
    WHERE table_schema = '{SCHEMA}' 
    AND table_type = 'BASE TABLE';
""")

tables = [t[0] for t in cur.fetchall()]

print("Tablas encontradas:", tables)

# 2. Exportar cada tabla a CSV
for table in tables:
    df = pd.read_sql(f'SELECT * FROM {SCHEMA}."{table}"', conn)
    df.to_csv(f"{OUTPUT_DIR}/{table}.csv", index=False)
    print(f"Exportada: {table}")

cur.close()
conn.close()
```
